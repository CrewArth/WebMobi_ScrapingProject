### **WEBSITE 1 : ALLEVENTS.IN**

```python
#Import the Required Libraries

import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = 'https://allevents.in/mumbai/all?ref=cityselect-eventlist-nearby'
res = requests.get(url)
status = res.status_code
print(status)
```

```python
#Contains Raw Html Data.

doc = res.text
# print(doc)
```

```python
#Using Beautiful Soup and Parsed Data.

soup = BeautifulSoup(doc, 'html.parser')
#print(soup)
```

```python
#We will fetch Title

titleList = []
titles = soup.find_all('h3', class_ = '')

for title in titles:
#   print(title.text)
#   title = title.text
    titleList.append(title.text.strip())


titleList
```

```python
#Fetching Locations of Events.

locationList = []

location_divs = soup.find_all('div', class_='subtitle')

# Extract the text from each div
locations = [div.get_text(strip=True) for div in location_divs]

# Print the locations
for location in locations:
    #print(location)
    locationList.append(location)


locationList
```

```python
#Fetching Dates for the Events.

dateList = []

date_divs = soup.find_all('div', class_='date')

dates = [div.get_text(strip=True) for div in date_divs]

for date in dates:
    #print(location)
    dateList.append(date)

dateList
```

```python
#Fetching Ticket Price for the Events.

price_spans = soup.find_all('span', class_='price')

# Initialize an empty list to store cleaned prices
priceList = []

# Extract the text from each span, clean it, then append to the list
for span in price_spans:
    price = span.get_text(strip=True).replace('\n', '').replace('\t', '')
    priceList.append(price)

# Print the cleaned prices
for price in priceList:
    print(price)
```

```python
#We are Fetching Selling Position

positionList = []

positions = soup.find_all('span', class_='c-ribbon')

for position in positions:
#   print(title.text)
#   title = title.text
    positionList.append(position.text.strip())

positionList
```

```python
#Compiling all the data into CSV.

FinalData = {
    'Title' : titleList,
     'Location' : locationList,
    'Date' : dateList,
   'Price' : priceList,
    'Position' : positionList
}

df = pd.DataFrame.from_dict(FinalData, orient='index')
df.transpose()
```

```python
df.to_csv('EventListing.csv')
```

```python
#Fetching Dates of the Events

date_spans = soup.find_all('span')

# Regular expression to match date-like patterns
date_pattern = re.compile(r'\b(?:\w+ \d{1,2}, \d{4})\b')

# Extract the text from each span tag and filter out non-date text
datesList2 = []
for span in date_spans:
    text = span.get_text(strip=True)
    if date_pattern.match(text):
        datesList2.append(text)

datesList2
```

```python
import requests
from bs4 import BeautifulSoup
import csv

# URL of the Eventbrite website to scrape
url = 'https://www.eventbrite.com/d/online/events/'

# Send a GET request to the URL
response = requests.get(url)
if response.status_code == 200:
    html = response.text
    soup = BeautifulSoup(html, 'html.parser')

    # Find all event listings
    events = soup.find_all('div', class_='search-event-card-wrapper')

    # List to store event data
    event_data = []

    # Extract event details
    for event in events:
        # Extract event title
        event_name_tag = event.find('div', class_='eds-event-card-content__content__principal')
        event_name = event_name_tag.get_text().strip() if event_name_tag else 'No event name found'

        # Extract event date
        event_date_tag = event.find('div', class_='eds-event-card-content__sub-content')
        event_date = event_date_tag.get_text().strip() if event_date_tag else 'No date found'

        # Extract event link
        event_link_tag = event.find('a', class_='eds-event-card-content__action-link')
        event_link = event_link_tag['href'] if event_link_tag else 'No link found'

        # Extract event price
        event_price_tag = event.find('div', class_='DiscoverVerticalEventCard-module__priceWrapper___2_JrI')
        event_price = event_price_tag.get_text().strip() if event_price_tag else 'Free'

        # Extract organizer name and followers
        organizer_name = "No organizer found"
        followers = "No followers found"
        organizer_tag = event.find('div', class_='card-text--truncated__one')
        if organizer_tag:
            organizer_link = organizer_tag.find('a')
            if organizer_link:
                organizer_name = organizer_link.get_text().strip()
                organizer_page = requests.get(organizer_link['href'])
                if organizer_page.status_code == 200:
                    organizer_soup = BeautifulSoup(organizer_page.text, 'html.parser')
                    followers_tag = organizer_soup.find('div', class_='js-followers-count')
                    if followers_tag:
                        followers = followers_tag.get_text().strip()

        # Append event details to the list
        event_data.append([event_name, event_date, event_price, organizer_name, followers, event_link])

    # Print event data (optional)
    for event in event_data:
        print(event)
else:
    print("Failed to retrieve the webpage")

# Save the data to a CSV file
csv_filename = 'eventbrite_events.csv'
with open(csv_filename, 'w', newline='', encoding='utf-8') as file:
    writer = csv.writer(file)
    writer.writerow(["Event Name", "Event Date", "Event Price", "Organizer Name", "Followers", "Event Link"])
    writer.writerows(event_data)

print(f"Data has been written to {csv_filename}")
```
